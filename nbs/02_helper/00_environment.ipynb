# Environment

> Exports the current conda environment to `01_Figures/SerotoninEthomics.yml`.
> Run this notebook whenever packages are added or updated.

```python
import subprocess, pathlib, sys
```

## Export environment

```python
yml_path = pathlib.Path('SerotoninEthomics.yml')

result = subprocess.run(
    ['conda', 'env', 'export', '--no-builds'],
    capture_output=True, text=True
)

if result.returncode != 0:
    raise RuntimeError(f'conda env export failed:\n{result.stderr}')

yml_path.write_text(result.stdout)
print(f'Exported to {yml_path.resolve()}')
print(f'{len(result.stdout.splitlines())} packages recorded')
```

## Key package versions

```python
key_packages = [
    'espresso', 'esploco', 'dabest',
    'numpy', 'pandas', 'scipy',
    'matplotlib', 'seaborn', 'scikit-learn',
]

import importlib.metadata
print(f'Python {sys.version.split()[0]}\n')
for pkg in key_packages:
    try:
        v = importlib.metadata.version(pkg)
        print(f'  {pkg:<20} {v}')
    except importlib.metadata.PackageNotFoundError:
        print(f'  {pkg:<20} not found')
```
